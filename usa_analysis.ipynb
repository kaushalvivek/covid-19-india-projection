```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import wilcoxon, ttest_ind, normaltest, shapiro
from scipy.stats import pearsonr
```

```python
deaths = pd.read_excel("./Spreadsheets/USA_masterSheet.xlsx",sheet_name="State_wise",index_col=0)
pop_density = pd.read_excel("./Spreadsheets/USA_masterSheet.xlsx",sheet_name="Population",index_col=0)
asthma = pd.read_excel("./Spreadsheets/USA_masterSheet.xlsx",sheet_name="Asthma",index_col=0)
age_dist = pd.read_excel("./Spreadsheets/USA_masterSheet.xlsx",sheet_name="Age_distribution",index_col=0)
lit_level = pd.read_excel("./Spreadsheets/USA_masterSheet.xlsx",sheet_name="Literacy Level",index_col=0)
hdi = pd.read_excel("./Spreadsheets/USA_masterSheet.xlsx",sheet_name="HDI",index_col=0)
diabetes = pd.read_excel("./Spreadsheets/USA_masterSheet.xlsx",sheet_name="Diabetes",index_col=0)
hiv = pd.read_excel("./Spreadsheets/USA_masterSheet.xlsx",sheet_name="HIV",index_col=0)
```

```python
print(deaths)
```

```python
df1 = deaths.merge(asthma,left_index=True,right_index=True)
df2 = df1.merge(pop_density,left_index=True,right_index=True)
df2 = df2.merge(age_dist,left_index=True,right_index=True)
df2 = df2.merge(lit_level,left_index=True,right_index=True)
df2 = df2.merge(hdi,left_index=True,right_index=True)
df2 = df2.merge(diabetes,left_index=True,right_index=True)
df2 = df2.merge(hiv,left_index=True,right_index=True)
```

```python
print(df2)
df2.to_excel("USA_consolidated.xlsx")
```
